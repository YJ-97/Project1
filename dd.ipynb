```python
trade_dependence = pd.read_csv('./datasets/trade_dependence.csv', encoding='CP949')
trade_dependence
```

```python
foreign_exchange_reserves = pd.read_csv('./datasets/foreign_exchange_reserves.csv', encoding='CP949')
foreign_exchange_reserves
```

```python
Debt_to_GDP_ratio = pd.read_csv('./datasets/Debt_to_GDP_ratio.csv', encoding='CP949')
Debt_to_GDP_ratio.columns = ['국가별', '시점', 'Debt_to_GDP_ratio']
Debt_to_GDP_ratio
```

```python
Current_Account = pd.read_csv('./datasets/Current_Account.csv', encoding='CP949')
Current_Account
```

```python
lists = [major, trade_dependence,foreign_exchange_reserves, Debt_to_GDP_ratio, Current_Account]

from functools import reduce

all = reduce(lambda left, right: pd.merge(left, right, on=['국가별', '시점'], how='left'), lists)
all.to_csv('./datasets/all.csv', encoding='CP949')
```
